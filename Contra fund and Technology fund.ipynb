```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split

def ConvertDateintoDay_Month_Year(df):
    #newdf = pd.DataFrame()
    df.Date = pd.to_datetime(df.Date)
    #df["Nav"] = df.Nav
    df["day"] = df.Date.dt.day
    df["month"] = df.Date.dt.month
    df["year"] = df.Date.dt.year
    df["day_of_week"] = df.Date.dt.dayofweek
    df = df.drop(columns=["Unnamed: 0","Date"],axis=1)
    return df
```

```python
directory =  os.path.join(os.getcwd())
file = r"SBI Contra fund featured data.csv"
df = pd.read_csv(os.path.join(directory,file))
df = ConvertDateintoDay_Month_Year(df)
print(df)
```

```python
newdf = pd.get_dummies(df, columns=['day_of_week', 'month'])
```

```python
newdf
```

```python
mm = preprocessing.MinMaxScaler()
```

```python
col_names = list(newdf.columns)
scaled_df = mm.fit_transform(newdf)
scaled_df = pd.DataFrame(scaled_df,columns = col_names)
scaled_df
```

```python
y = scaled_df["Open"]
X = scaled_df.drop(columns=["Open","risk_free_rate"],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
def train_model(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
    
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    regr_y_pred = regr.predict(X_test)
    mae = mean_absolute_error(y_test, regr_y_pred)
    print("Linear regression:")
    print('MAE:', mae)
    print("Model Score : ",regr.score(X_test, y_test))
    print("------------------------------------")
    
    rnd = RandomForestRegressor()
    rnd.fit(X_train, y_train)
    rnd_y_pred = rnd.predict(X_test)
    mae = mean_absolute_error(y_test, rnd_y_pred)
    print("Random Forest Regression")
    print('MAE:', mae)
    print("Model Score: ",rnd.score(X_test, y_test))
    print("-------------------------------------")
    
    model = Pipeline([('scaler', StandardScaler()), ('svr', SVR())])
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    print("Support Vector Regressor")
    print('MAE:', mae)
    print("Model Score: ",model.score(X_test, y_test))
    
    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

train_model(X,y)
```
