```python
def ConvertDateintoDay_Month_Year(df):
    newdf = pd.DataFrame()
    df.Date = pd.to_datetime(df.Date)
    newdf["FundName"] = df.FundName
    newdf["Nav"] = df.Nav
    newdf["day"] = df.Date.dt.day
    newdf["month"] = df.Date.dt.month
    newdf["year"] = df.Date.dt.year
    return newdf
def ConvertDateintoDay_Month_Year_(df):
    newdf = pd.DataFrame()
    df.Date = pd.to_datetime(df.Date)
    newdf["Nav"] = df.Nav
    newdf["day"] = df.Date.dt.day
    newdf["month"] = df.Date.dt.month
    newdf["year"] = df.Date.dt.year
    return newdf
```

```python
def OptimiseDataframe(df,fundname):
    l = len(df["Date"].tolist())
    fund = [fundname]*l
    newdf = pd.DataFrame()
    newdf["FundName"] = fund
    newdf["Date"] = df.Date
    newdf["Nav"] = df.Open
    newdf.dropna()
    return newdf
```

```python
def CalculateReturn(start,end):
    returns = (end-start)/start
    returns = returns*100
    return returns
    
    
```
