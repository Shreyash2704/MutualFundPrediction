```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
```

```python
def ConvertDateintoDay_Month_Year(df):
    #newdf = pd.DataFrame()
    df.Date = pd.to_datetime(df.Date)
    #df["Nav"] = df.Nav
    df["day"] = df.Date.dt.day
    df["month"] = df.Date.dt.month
    df["year"] = df.Date.dt.year
    df["day_of_week"] = df.Date.dt.dayofweek
    df = df.drop(columns=["Unnamed: 0","Date"],axis=1)
    return df

def OptimiseDataframe(df):
    newdf = pd.DataFrame()
    newdf["Date"] = df.Date
    newdf["Nav"] = df.Open
    newdf = newdf.dropna()
    return newdf
```

```python
directory =  os.path.join(os.getcwd())
file = r"SBI Contra fund featured data.csv"
df = pd.read_csv(os.path.join(directory,file))
#newdf = OptimiseDataframe(df)
#print(newdf)
new_df = ConvertDateintoDay_Month_Year(df)
print(new_df)
```

```python
day_month_dummies = pd.get_dummies(new_df, columns=['day_of_week', 'month'])
```

```python
data_to_be_trained = day_month_dummies.drop(columns=["day"],axis=1)
```

```python
new_df = X[60::]
new_df = new_df.reset_index()
historic_scaled_df = historic_scaled_df.reset_index()
new_dataset = pd.concat([new_df, historic_scaled_df], axis=1)
```

```python
new_dataset = new_dataset.drop(columns=["index"],axis=1)
```

```python
new_y = y[60::]
new_y
```

```python
def train_model(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
    
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    regr_y_pred = regr.predict(X_test)
    mae = mean_absolute_error(y_test, regr_y_pred)
    print("Linear regression:")
    print('MAE:', mae)
    print("Model Score : ",regr.score(X_test, y_test))
    print("------------------------------------")
    
    rnd = RandomForestRegressor()
    rnd.fit(X_train, y_train)
    rnd_y_pred = rnd.predict(X_test)
    mae = mean_absolute_error(y_test, rnd_y_pred)
    print("Random Forest Regression")
    print('MAE:', mae)
    print("Model Score: ",rnd.score(X_test, y_test))
    print("-------------------------------------")
    
    model = Pipeline([('scaler', StandardScaler()), ('svr', SVR())])
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    print("Support Vector Regressor")
    print('MAE:', mae)
    print("Model Score: ",model.score(X_test, y_test))
    
    
```

```python
train_model(new_dataset,new_y)
```
